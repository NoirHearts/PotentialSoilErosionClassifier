```python
import pandas as pd
```

```python
luid = pd.read_csv("/content/luid.csv") #load test set
luid_train = pd.read_csv("/content/luid_train.csv") #load train set


x = luid_train.drop(["Potential Soil Erosion"], axis = 1) # get x variables
y = luid_train.drop(["Slope1", "Elevation1", "Land Cover1"], axis = 1) #get y variable
luid = luid.dropna() #remove all null values
```

```python
# map values and categorize them by numbers
d1 = {'Gentle': 1, 'Level': 1, 'Undulating': 1,'Gentle Sloping': 1,'Moderately Steep': 2, 'Rolling': 2, 'Steep': 3, 'Very Steep': 3}
x['Slope1'] = x['Slope1'].map(d1)
luid['Slope1'] = luid['Slope'].map(d1)

d2 = {'Low': 1, 'Middle': 2, 'High': 3}
x['Elevation1'] = x['Elevation1'].map(d2)
luid['Elevation1'] = luid['Elevation'].map(d2)

d3 = {'Built-up Area': 1,'Built-up': 1, 'Closed Canopy': 2, 'Closed Can': 2, 'Cultivated': 3, 'Grassland': 4, 'Open Canopy': 5}
x['Land Cover1'] = x['Land Cover1'].map(d3)
luid['Land Cover1'] = luid['Land Cover'].map(d3)

d4 = {'Low': 1, 'Moderate': 2, 'Severe': 3, 'Very Severe': 4}
y['Potential Soil Erosion'] = y['Potential Soil Erosion'].map(d4)
```

```python
print(luid) #check dataframe
```

```python
# use decision tree as ml model
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

clf = DecisionTreeClassifier()
clf = clf.fit(x,y)
```

```python
luid_test = luid.iloc[:, [14,15,16]] # get x variables of test set
luid_test
```

```python
# predict y from test variables
y_pred = clf.predict(luid_test)
y_pred
```

```python
# fixing the table columns and adding results to the table
luid['Potential Soil Erosion'] = y_pred.tolist()
luid = luid.drop(["Slope1", "Elevation1", "Land Cover1"], axis = 1)

id4 = dict(zip(d4.values(), d4.keys()))
luid['Potential Soil Erosion'] = luid['Potential Soil Erosion'].map(id4)
print(luid)
```
